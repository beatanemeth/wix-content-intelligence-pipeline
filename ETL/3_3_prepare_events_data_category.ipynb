# Configurations

```python
# Define relative paths
INPUT_FILE_EVENTS = "../data_temp/events_2_speakers_intermediate.json"
OUTPUT_FOLDER = "../data_temp/"
OUTPUT_FILE = os.path.join(OUTPUT_FOLDER, "events_3_category.json")

# Simple check to verify the file is where we think it is
if os.path.exists(INPUT_FILE_EVENTS):
    print(f"✅ Setup complete. Input file found: {INPUT_FILE_EVENTS}")
else:
    print(f"❌ Warning: Input file NOT found at {INPUT_FILE_EVENTS}")

# Ensure the output directory exists
if os.path.exists(OUTPUT_FOLDER):
    print(f"✅ Setup complete. Output folder found: {OUTPUT_FOLDER}")
else:
    os.makedirs(OUTPUT_FOLDER)
    print(f"✅ Created Output folder: {OUTPUT_FOLDER}")
```

# Load raw data

```python
try:
    events_df = pd.read_json(INPUT_FILE_EVENTS)
    print(f"✅ Successfully loaded {len(events_df)} records.")
    
    display(events_df.head(2)) 
    
    print("\n📑 Available columns:", *events_df.columns, sep="\n")
    
except FileNotFoundError:
    print(f"Error: The file {INPUT_FILE_EVENTS} was not found.")
```

# Check whether all items have value for `category`

```python
# 1. Display all events without a category (where category is NaN/null)
events_without_category_df = events_df[events_df['category'].isnull()]
print(f"✅ Found {len(events_without_category_df)} events without a category.")
display(events_without_category_df[['id', 'title', 'date', 'speaker', 'category', 'url']])

# 2. Display all unique category names
# First, drop nulls, then convert to string to handle potential mixed types,
# then split by comma and explode to get individual category items,
# finally strip whitespace and get unique values.
unique_categories = (
    events_df['category']
    .dropna()
    .astype(str)
    .str.split(',')
    .explode()
    .str.strip()
    .unique()
)

# Convert numpy array to list for cleaner printing

print("\n" + "="*50 + "\n")
print(f"✅ Found {len(unique_categories)} unique categories:")
print(*unique_categories, sep="\n")

```

# Export to JSON

```python
# force_ascii=False is crucial for keeping Hungarian characters like ő, ú, é
if pd.api.types.is_datetime64_any_dtype(events_df['date']):
    events_df['date'] = events_df['date'].dt.strftime('%Y-%m-%d')

# date_format='iso' ensures dates are readable strings
# categories will be included as long as they are in the events_df
events_df.to_json(
    OUTPUT_FILE, 
    orient='records', 
    indent=4, 
    force_ascii=False, 
    date_format='iso'
)

print(f"✅ Success! Cleaned data is saved to: {OUTPUT_FILE}")
```
